```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

# Load the face and eye classifiers
face_classifier = cv2.CascadeClassifier("/content/haarcascade_frontalface_default.xml")
eye_classifier = cv2.CascadeClassifier("/content/haarcascade_eye.xml")

# Load the image and convert it to grayscale
image = cv2.imread("/content/th (6).jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces in the image
faces = face_classifier.detectMultiScale(gray, 1.3, 5)

# Check if no faces are found
if len(faces) == 0:
    print("No faces found")

# Process each detected face
for (x, y, w, h) in faces:
    # Draw a rectangle around each face
    cv2.rectangle(image, (x, y), (x + w, y + h), (127, 0, 255), 2)

    # Region of interest (ROI) for eyes detection
    roi_gray = gray[y:y + h, x:x + w]
    roi_color = image[y:y + h, x:x + w]

    # Detect eyes in the region of interest
    eyes = eye_classifier.detectMultiScale(roi_gray)

    # Draw rectangles around detected eyes
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)

# Show the final image with faces and eyes detected
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
face_classifier=cv2.CascadeClassifier("/content/haarcascade_frontalface_default.xml")
eye_classifier=cv2.CascadeClassifier("/content/haarcascade_eye.xml")
image=cv2.imread("/content/th (6).jpg")
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
faces=face_classifier.detectMultiScale(image,1.3,5)
if faces is ():
  print("no face found")

for (x,y,w,h) in faces:
  cv2.rectangle(image,(x,y),(x+w,y+h),(127,0,255),2)
  cv2_imshow(image)
  cv2.waitKey(0)
  roi_gray=gray[y:y+h,x:x+w]
  roi_color=image[y:y+h,x:x+w]
  eyes=eye_classifier.detectMultiScale(roi_gray)
  for (cx,cy,cw,ch) in eyes:
    cv2.rectangle(roi_color,(cx,cy),(cx+cw,cy+ch),(255,255,0),2)
    cv2_imshow(image)

cv2.destroyAllWindows()
```
